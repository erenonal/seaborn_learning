# Dogrusal İlişkinin Gösterilmesi

#### geçenki kodda scotterplot ile yaptığımız saçılım grafiğini şimdi doğrusal olarak ifade etmeye çalışıcaz.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
#yukarda görüleceği üzere matplotlibe içerisinedki tek bir fonksiyonu import ettik komple matplotulibi değil
```

```python
# doğrusal bir çizgi koyacağımız için lineer ifadeyi kullanıyoruz lmplot şeklide
```

```python
sns.lmplot( x = "total_bill", y = "tip", data =df);
```

```python
#doğru etrafındak şeffaf mavi kısım ise sapmaları ifade ediyo verilerin yoğun olduğu yerde bu tabaka görünmezken verilerin uzak olduğu yerlerde sağmalardan dolayı böyle bir şekil oluşuyor.
```

```python
sns.lmplot( x = "total_bill", y = "tip", hue = "smoker", data =df);
```

```python
#aradaki eğim farkından yorum yapabiliriz. mesela sigara içmeyen insanlarda hesap arttıkça bahşişin artar yorumunu yapabilirken, içmeyen insanlarda bu tezi savunamayız.
# bu tabiki bir tez sadece buna bakılarak böyle bi yorum yapılması doğru olmaz
```

```python
sns.lmplot( x = "total_bill", y = "tip", hue = "smoker", col = "time",  data =df);
```

```python
# yukarda col fonksiyonu column yani sütun boyutunda zaman değişkeninin ekledi. lunch ve dinner olarak. burdan yola çıkarak şunu varsayabiliriz
# öğlen sigara içip içmeme durumu çok farketmezken akşam hesapla beraber ödenen bahşiş miktarının da artmsında sigara içip içmeme durumunun fark ettiğini söyleyebiliriz.
```

```python
#bir boyut daha eklemek istersek bunu da row yani satır boyutunda ekleyebiliriz .
```

```python
sns.lmplot( x = "total_bill", y = "tip", hue = "smoker", col = "time", row = "sex",  data =df);
```

```python
# bunları biz veriyi tanımak için yapıoruz.bunu algoritmaya gönderdiğimizde mantıklı sonuç alıp almadığımız hakkınd bir bilgi edinmemize yardımcı olur.
```

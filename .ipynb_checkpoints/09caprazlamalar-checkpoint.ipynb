## Çaprazlamalar

```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
#sürekli edğişkenin grafiğini oluştur dediğimizde bize böyle bi yoğunluk grafiği verdi
```

```python
sns.kdeplot(df.price, shade = True);
```

```python
#bizim gördüğümüz bu mavi alanı oluşturan nedir. grafiğin bizen sakladığı şeyler nedir bunu sorguluyoruz. çaprazlamalar ile nasıl olmuş sorusuna cevap arıyoruz
```

```python
#aşağıdaki kdeplot ile yoğunluğu görselleştirirken FacetGrid içerisindeki cut kategorik değişkeni ile mapl'leyerek boyut eklemek istiyoruz
```

```python
(sns
 .FacetGrid(df,
                    hue = "cut",
                    height = 5,
                    xlim = (0, 10000))
 .map(sns.kdeplot, "price", shade = True)
 .add_legend()
); 
```

```python
sns.catplot(x = "cut", y = "price", hue = "color", kind = "point", data = df)
```

```python
#bir sayısal price, iki tane kategorik değişkenimiz var color ve cut
#fiyatın farklı seviylereinde iki kategorik değişkenin etkisini görebiliyoruz.
#cutun içindeki ideale baktığımızda  color kategorilerinin dağılıyor olması price kısmında farklı olması bir bilginin taşındığı anlamına geliyor
#amacımız bu farklılığı farklı değişkenlerle ayırt edilerek açıklamaya çalışmaktır.
#değerler yine 0-6k arasında bunun nedeni farklı 3 değişkenin birbiri ile aggregate edilmiş olması yani toplulaştırılmış olması iki kere groupby yapıldıktan sonra elde edilecek rotalama değerler
#^BOYUTLU GRAFİK BU 
```

# Scatterplot Matrisi

```python
#scatterplot sayısal değişkenlerin arsındaki ilişkiyi ifade eden bir saçılım grafiği idi.
#matris boyutunda bu ilişki ifade edildiğinde bir scatterplot matris oluşturmuş oluyoruz.
```

```python
import seaborn as sns
iris = sns.load_dataset("iris")
df = iris.copy()
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
sns.pairplot(df);
```

```python
#yukarda görüldüğü üzere veri setindeki 4 farklı değişkenin birbiri ile ilişkileri ortaya konumuştur.
```

```python
#scatterplot görsellerinde eğer oluşturulmuş olan grafi bir toz bulutu şeklinde ise veya yapısal bir formu yoksa bu o iki değişken arasında bir ilişki olmadığı 
#anlamına gelir
```

```python
#oluşan scatterplot grafiklerinde gözlem noktaları iki farklı köşeye kümelendiyse bu çaprazlamalar sonucu oluşan bu ifadeleri oluşturan bazı alt gruplar olduğunu 
#ve bunların incelenmesi gerektiğini bildirir. aşağıda bu duruma örnek verilmiştir.
```

```python
sns.pairplot(df,hue = "species");
```

```python
#1. grafikte 3 tepecik var demek ki 3 farklı tür var diyebiliriz.
```

```python
#grafiklerdeki şekilleri değiştirmek içinse aşağıdaki yolu izlememiz gerekiyor
#markers = [""]şeklinde
```

```python
sns.pairplot(df,hue = "species", markers = ["o","s","D"]);
```

```python
#her graiğe bir doğru eklemek için de king = "reg şeklinde yapabiliriz"
```

```python
sns.pairplot(df,kind = "reg");
```

```python
#aşağıda ise eğriler olarak 
```

```python
sns.pairplot(df,kind = "reg", hue ="species");
```

# Heatmap

### Veri Seti Hikayesi


```python
import seaborn as sns
flights = sns.load_dataset("flights")
df = flights.copy()
df.head()
```

```python
df.shape
```

```python
df["passengers"].describe().T
```

```python
df[["passengers"]].describe().T
```

```python
#pivot table yaparken aşağıda ayları x yılları y ekesnine koyarak daha okunaklı bir tablo haleine koyduk
```

```python
df = df.pivot("month", "year", "passengers")
df
```

```python
sns.heatmap(df);
```

```python
sns.heatmap(df, annot = True, fmt = "d");
```

```python
sns.heatmap(df, annot = True, fmt = "d", linewidths = 0.5);
```

```python
sns.heatmap(df, annot = True, fmt = "d", linewidths = 0.5, cbar = False);
```
